```python
import pandas as pd
import json
import urllib.request
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 2)
```

```python
#Number codes for all counties in PA 
nums = ['045', '119', '127', '043', '129', '053', '089', '101', '013', '021', '011', '061', '073', '087', '041', '003', '109', '115',
 '039', '049', '057', '067', '107', '131', '001', '005', '035', '025', '027', '097', '095', '023', '121', '007', '017',
 '069', '083', '051', '091', '059', '075', '079', '093', '037', '019', '065', '071', '099', '105', '117', '055', '009',
 '111', '133', '015', '029', '063', '125', '103', '077', '031', '033', '081', '123', '047', '085', '113']
```

```python
all_data = []

for i in nums:
    # open a connection to a URL using urllib
    url = 'https://api.census.gov/data/2018/acs/acs5?get=NAME,B01002_001E,B01002_002E,B01002_003E,B01002B_001E,B01002B_002E,B01002B_003E,B01002C_001E,B01002C_002E,B01002C_003E,B01002D_001E,B01002D_002E,B01002D_003E,B01002E_001E,B01002E_002E,B01002E_003E,B01002H_001E,B01002H_002E,B01002H_003E,B01002I_001E,B01002I_002E,B01002I_003E&for=block+group:*&in=state:42+county:' + i + '&key=0b1e1ad42a064e1b73b75c69016f2965a7b1d35d'
    webUrl  = urllib.request.urlopen(url).read()
    data = json.loads(webUrl)

    for row in data:
        if row == data[0]:
            continue        
        full_row = []
        full_row=row[0].split(',') + row[1:len(row)]
        
        #append each county to full data set
        all_data.append(full_row)

```

```python
df = pd.DataFrame(all_data, columns = ['block_group', 'census_tract', 'county', 'state','MEDIAN AGE BY SEX-Total', 'MEDIAN AGE BY SEX-Male', 'MEDIAN AGE BY SEX-Female', 'MEDIAN AGE BY SEX-Black-Total', 'MEDIAN AGE BY SEX-Black-Male', 'MEDIAN AGE BY SEX-Black-Female', 'MEDIAN AGE BY SEX-Amer_Alaska_Native-Total', 'MEDIAN AGE BY SEX-Amer_Alaska_Native-Male', 'MEDIAN AGE BY SEX-Amer_Alaska_Native-Female', 'MEDIAN AGE BY SEX-Asian-Total', 'MEDIAN AGE BY SEX-Asian-Male', 'MEDIAN AGE BY SEX-Asian-Female', 'MEDIAN AGE BY SEX-Hawaiian_Pac_Islander-Total', 'MEDIAN AGE BY SEX-Hawaiian_Pac_Islander-Male', 'MEDIAN AGE BY SEX-Hawaiian_Pac_Islander-Female', 'MEDIAN AGE BY SEX-White_Non_Hisp-Total', 'MEDIAN AGE BY SEX-White_Non_Hisp-Male', 'MEDIAN AGE BY SEX-White_Non_Hisp-Female', 'MEDIAN AGE BY SEX-Hispanic_Latino-Total', 'MEDIAN AGE BY SEX-Hispanic_Latino-Male', 'MEDIAN AGE BY SEX-Hispanic_Latino-Female', 'state_code', 'county_code', 'tract_code', 'block_group_code'])
```
